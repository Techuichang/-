```python
df_new['markx']=0


def fuc_x_i(i,m,x=0):#寻找该点后的第一个相反分型的索引,i为该点索引，设置m为相反的顶底分型记号,****注意****从该点以后，不含该点的符合mark值条件的作为第一个分型
    global df_new
    return df_new.loc[ (df_new.index >i) & (df_new['markf'] == m)].head(x+1).index.values[x]


def fuc_tzxl(i,x,n=0): #返回一个列表，第一个元素为较大值，第二个为较小值  ***注意*** 参数i,x,n意思为从i点开始的（不含该点，比如i是底分型位置，那么）的，对应x值的第n+1（n默认为0）个分型（x为-1是顶分型）的位置
    global df_new
    return ([max(df_new.loc[fuc_x_i(df_new,i,-x,n)].high, df_new.loc[fuc_x_i(df_new,i,x,n)].low) ,
              min(df_new.loc[fuc_x_i(df_new,i,-x,n)].high, df_new.loc[fuc_x_i(df_new,i,x,n)].low)])

def fuc_txzl_fx(x,t1,t2,t3): 
    if x==-1:
        if ((t2[0]>t1[0]) and (t2[0]>t3[0])) and ((t2[1]>t1[1]) and (t2[1]>t1[1])):
            return True    #分型出现
        else:
            return False
    if x==1:
        if ((t2[0]<t1[0]) and (t2[0]<t3[0])) and ((t2[1]<t1[1]) and (t2[1]<t1[1])):
            return True    #分型出现
        else:
            return False        

def fuc_tzxl_hb(x,k,j):
    if x==-1:
        return [max(k[0],j[0]),max(k[1],j[1])]
    if x==1:
        return [min(k[0],j[0]),min(k[1],j[1])]

def fuc_tzxl_qk(t1,t2):
    if (t1[0]<t2[1]) or (t1[1]>t2[0]):
        return True
    else:
        return False

    
def fuc_tzxl_mv(t2,t2_index):#线段移动
    s = 0
    i = t3_index
    
    t3 = fuc_tzxl(i,x,s)      #新的第三特征序列为，i(t3_index所在的底)以后，从第s（1）个顶开始的特征序列
    t3_index = fuc_x_i(i,x,s) #i(t3_index所在的底)以后的第s(1)个底分型位置作为t3_index的位置   
    
    flag = 1
    while flag ==1:    #条件循环结束后 t2 和t3 必然不是重合
        if (t2[0]>t3[0] and t2[1]<t3[1]) or (t2[0]<t3[0] and t2[1]>t3[1]):
            t2 = fuc_tzxl_hb( x，t2,t3 )    #以x为参数，t2修订为t2和t3合并
            t2_index = fuc_x_i(i,x,s)       #合并后，将t2_index修订为原t3_index
            s= s+1
            t3 = fuc_tzxl(i,x,s)            #原t3修订为，i(底的位置)以后，从第s（2）个顶开始的特征序列
            t3_index = fuc_x_i(i,x,s)       #原t3_index的位置修订为i以后的第s(2)个底分型位置
        else:
            flag = -1                       #t2和t3并非包含关系时，循环条件关闭
    return ([t3,t3_index])                   
    
def fuc_tzxl_new(i,x):
    global df_new
    flag = 1    #用于判断 特征序列1和特征序列2是否需要合并
    s = 0       #用于辅助判断 特征序列合并后确定下一个特征序列
    t1 = fuc_tzxl(i,x,s)      #i(底的位置)以后，从第s（1）个顶开始的特征序列
    t1_index = fuc_x_i(i,x,s)  #i以后的第s(1)个底分型位置作为t1_index的位置
    s = s + 1                #此时s=1
    t2 = fuc_tzxl(i,x,s)     #i(底的位置)以后，从第s（2）个顶开始的特征序列
    t2_index = fuc_x_i(i,x,s)#i以后的第s(2)个底分型位置作为t2_index的位置    
    
    while flag ==1:    #条件循环结束后 t1 和t2 必然不是重合
        if (t1[0]>t2[0] and t1[1]<t2[1]) or (t1[0]<t2[0] and t1[1]>t2[1]):
            t1 = fuc_tzxl_hb( x，t1,t2 )#以x为参数，t1修订为t1和t2合并
            t1_index = t2_index         #合并后，将t1_index修订为原t2_index
            s= s+1                      #在第一次合并后，s为2
            t2 = fuc_tzxl(i,x,s)        #原t2修订为，i(底的位置)以后，从第s（3）个顶开始的特征序列
            t2_index = fuc_x_i(i,x,s)   #原t2_index的位置修订为i以后的第s(3)个底分型位置
        else:
            flag = -1                   #t1和t2并非包含关系时，循环条件关闭    
    return ([t1,t1_index,t2,t2_index])


def fuc_xian(df):#从上升线开始
    global df_new
    
    i = df_new.loc[df_new['markf']==-1]['low'].idxmin()#寻找最低点的底分型
    
    x = -1      #fuc_x_i 参数 ，用于控制确定特征序列的方向
    
    tt = fuc_tzxl_new(i,x)    #以下功能块用于生成三个标准元素
    t1 = tt[0]
    t1_index = tt[1]
    t2 = tt[2]
    t2_index = tt[3]   
    i = t1
    tt = fuc_tzxl_new(i,x)
    t2 = tt[0]
    t2_index = tt[1]
    t3 = tt[2]
    t3_index = tt[3]


    while fuc_tzxl_fx(x,t1,t2,t3)!=True:#在当前三元素没有出现对应x参数的分型时，继续寻找下三个元素
        
        t1 = t2              #将第一特征序列修订为原第二特征序列
        t1_index = t2_index  #将第一特征序列起点修订为原第二特征序列起点

        t2 = t3              #将第二特征序列修订为原第三特征序列
        t2_index = t3_index  #将第二特征序列起点修订为原第三特征序列起点
        
        tt = fuc_tzxl_mv(t2,t2_index) #生成第三特征序列
        
        t3 = tt[0]           
        t3_index = tt[1]
        
        qk_f = fuc_tzxl_qk(t1,t2)           #给特征序列缺口是否存在标识qk_f赋值
    
        while fuc_tzxl_fx(x,t1,t2,t3)==True: #在当前三特征序列元素出现出现对应x参数的分型时
            if qk_f ==False:              #一二元素无缺口
                x = -x                     #寻找反向线段
                i = fuc_x_i(t1_index,-x)   #将第一元素的起始点的开始的下一个相反分型设为起始点
                df_new.loc[i,'markx'] = -x #在线段的终点位置的markx列的值标记出线段类别
                    
                tt = fuc_tzxl_new(i,x)    #以下功能块用于生成三个标准元素
                t1 = tt[0]
                t1_index = tt[1]
                t2 = tt[2]
                t2_index = tt[3]   
                i = t1
                tt = fuc_tzxl_new(i,x)
                t2 = tt[0]
                t2_index = tt[1]
                t3 = tt[2]
                t3_index = tt[3]
                
                break                      #生成三元素后跳出本循环进入上层循环
                

            if qk_f :                      #第一和第二元素***有特征序列缺口***，必须在相反分型确定后才能确定线段结束位置
                x = -x                     #寻找反向线段
                i = fuc_x_i(t1_index,-x)
                i_t = i #将第一次出现分型的位置记录在临时变量i_t t为temporary
                
                tt = fuc_tzxl_new(i,x)      #以下功能块用于生成三个标准元素
                t1 = tt[0]
                t1_index = tt[1]
                t2 = tt[2]
                t2_index = tt[3]   
                i = t1
                tt = fuc_tzxl_new(i,x)
                t2 = tt[0]
                t2_index = tt[1]
                t3 = tt[2]
                t3_index = tt[3]

                while fuc_tzxl_fx(x,t1,t2,t3)!=True:
                    t1 = t2              #将第一特征序列修订为原第二特征序列
                    t1_index = t2_index  #将第一特征序列起点修订为原第二特征序列起点

                    t2 = t3              #将第二特征序列修订为原第三特征序列
                    t2_index = t3_index  #将第二特征序列起点修订为原第三特征序列起点

                    tt = fuc_tzxl_mv(t2,t2_index) #生成第三特征序列

                    t3 = tt[0]
                    while fuc_tzxl_fx(x,t1,t2,t3)==True: #在当前三特征序列元素出现出现对应x参数的分型时

                            i = fuc_x_i(t1_index,-x)   #将第一元素的起始点的开始的下一个相反分型设为起始点
                            df_new.loc[i,'markx'] = -x #在线段的终点位置的markx列的值标记出线段类别

                            tt = fuc_tzxl_new(i,x)    #以下功能块用于生成三个标准元素
                            t1 = tt[0]
                            t1_index = tt[1]
                            t2 = tt[2]
                            t2_index = tt[3]   
                            i = t1
                            tt = fuc_tzxl_new(i,x)
                            t2 = tt[0]
                            t2_index = tt[1]
                            t3 = tt[2]
                            t3_index = tt[3]

                            break                      #生成三元素后跳出本循环进入上层循环                    



            
        
```
