```python
from selenium import webdriver
import time
import random

driver = webdriver.Chrome(r'C:\chromedriver')

import pandas as pd

a = pd.read_excel(r'C:\Users\Administrator\Downloads\上市公司列表.xlsx')

t=a[r'公司全称']
```

```python
def get():
    try:
        result = driver.find_element_by_xpath("/html/body/div[7]/section[1]/div[2]/div[2]/div[1]/div[1]/div/a").get_attribute("href")
    except:
        result = driver.find_element_by_xpath("/html/body/div[7]/section[1]/div[2]/div[2]/div[1]/div/div/a").get_attribute("href")
    return result
    
    
def go(t):

    driver.get('https://www.kanzhun.com/search/?q='+t)  #搜索页 可能进入验证码
    
    flag = 0
    
    while flag !=1 : #当flag不等于一时无法离开循环，实现验证码判定 除非发现页面有不是验证码页面元素，赋值flag为一，离开循环
        
        try:#如果进入验证码页面则无法找到相关元素，报错后进入except
            if driver.find_element_by_xpath("/html/body/div[1]/div/ul/li[1]/a").text == '综合': 
                flag=1  
        except:
            time.sleep(15) #进入验证码页面后手动输入验证码，完成后会自动跳转

    flag2 = 0 #当判断
    
    while flag2 != 1:###yanzhengma 
        time.sleep(1)
        try:
            result = get()   #在页面中寻找第一条或唯一搜索结果
            
            flag2 = 1
        except:
            time.sleep(1)
            if driver.find_element_by_xpath("/html/body/div[1]/div/ul/li[1]/a").text == '综合':
                return [t,False]
                
            
    try:
        driver.get(result)
    except:
        return [t,False]
    
    try:
        name = driver.find_element_by_xpath("/html/body/section[1]/div/div/div[2]/div[3]/div[1]/h1").text
    except:
        name = False
    
    try:
        score = driver.find_element_by_xpath("/html/body/section[1]/div/div/div[2]/div[2]/div[1]/div[2]").text
    except:
        score = False
    
    try:
        rate_num = driver.find_element_by_xpath("/html/body/div[2]/section[1]/section[2]/div[1]/div/dl/dd/p[2]/em").text
    except:
        rate_num = False
    
    try:
        fol_num = driver.find_element_by_xpath("/html/body/section[1]/div/div/div[2]/div[2]/div[2]/div[2]/span/em").text
    except:
        fol_num = False

    
    
    try:
        sg1 = driver.find_element_by_xpath("/html/body/div[2]/section[1]/div[2]/div[1]/div[1]").text
        sg2 = driver.find_element_by_xpath("/html/body/div[2]/section[1]/div[2]/div[2]/div[1]").text
        sg3 = driver.find_element_by_xpath("/html/body/div[2]/section[1]/div[2]/div[3]/div[1]").text
        l=[sg1,sg2,sg3]
    except:
        l = [0]
    try:
        sc1 = driver.find_element_by_xpath("/html/body/div[2]/section[1]/section[2]/div[1]/ul/li[1]/div/div/span").text
        sc2 = driver.find_element_by_xpath("/html/body/div[2]/section[1]/section[2]/div[1]/ul/li[2]/div/div/span").text
        sc3 = driver.find_element_by_xpath("/html/body/div[2]/section[1]/section[2]/div[1]/ul/li[3]/div/div/span").text
        s = [sc1,sc2,sc3]
    except:
        s = [0]
    return [t,name,score,rate_num,fol_num,l,s]
```

```python
for i in t:
    f = go(i)
    print(f)
```
